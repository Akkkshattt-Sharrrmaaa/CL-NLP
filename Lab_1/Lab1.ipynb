```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Salaries.csv')
data_df = pd.DataFrame(data)
```

```python
print(f"Shape of the dataset :{data_df.shape}")
```

```python
print("Overview of the dataset : ")
data_df.head()
```

```python
print("Statistical analysis of the dataset : ")
data_df.describe()
```

```python
data_df.isnull().sum()
```

```python
data_df['Location'].value_counts()
```

```python
print("Mean Salary:", round(data_df["Salary"].mean()))
print("Median Salary:", round(data_df["Salary"].median()))
print("Highest Salary:", round(data_df["Salary"].max()))
print("Lowest Salary:", round(data_df["Salary"].min()))
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = data_df["Location"], y = data_df["Salaries Reported"])
plt.show()
```

```python
plt.figure(figsize = (12, 6))
plt.ticklabel_format(style = 'plain')
sns.barplot(y = data["Salary"], x = data["Location"], palette = "tab20")
```

```python
plt.figure(figsize = (12, 6))
sns.scatterplot( x = "Rating", y = "Salaries Reported", data=data)
plt.show()
```
