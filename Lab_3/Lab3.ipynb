```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from matplotlib import pyplot as plt
from wordcloud import WordCloud
```

```python
file_path = './lab3_text.txt'
with open(file_path, 'r') as file:
    data = file.read()
data
```

```python
# performing word tokenization
workTokens = word_tokenize(data)
print(workTokens)
```

```python
# removing stop words
engStopwords = stopwords.words('english')
tokens_without_stopwords = [word for word in workTokens if word not in engStopwords]
tokens_without_stopwords
```

```python
# getting the frequency of words
fq = FreqDist(token.lower() for token in tokens_without_stopwords)
fq.most_common(10)
```

```python
x = []
y = []
for token in fq.most_common(10):
    x.append(token[0])
    y.append(token[1])
    
plt.plot(x, y)
```

```python
cloud = WordCloud(background_color='white').generate(data)
plt.figure(figsize = (15, 15))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
